# BÃ i 3

```python
img = cv2.imread('datdayaset//1.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
new_img = img.astype(np.float32)+50
img = np.clip(new_img, 0, 255)
img = img.astype(np.uint8)
plt.imshow(img)
plt.show()

new_img = img.astype(np.float32)-50-80
img = np.clip(new_img, 0, 255)
img = img.astype(np.uint8)
plt.imshow(img)
plt.show()
```
