```python
with open("data.txt", 'r') as f:
    data = f.read().replace('\n', ' ').lower()
    print("Câu 1:", data)

distinct_words = sorted(list(set(data.split(" "))))
print("Câu 2:", distinct_words)

print("Câu 3:")
mf, lfw = 0, []
lc, lcw = float('inf'), []

for word in distinct_words:
    count = data.split(' ').count(word)
    print(f"\"{word}\" in data is {count}")
    if count > mf:
        mf = count
        mfw = [word]
    elif count == mf:
        mfw.append(word)
    if count < lc:
        lc = count
        lcw = [word]
    elif count == lc:
        lcw.append(word)

print(f"\"{mfw}\" is/are most frequent word(s)")
print(f"\"{lcw}\" is/are least common word(s)")
```
